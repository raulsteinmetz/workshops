# Unsupervised Learning

## Separando espécies de flor por grupos de dados

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

Carregando os dados

```python
iris = load_iris()
```

Criando modelo de Machine learning (kmeans)

```python
kmeans = KMeans(n_clusters=3)
```

Treinando ele nos dados não rotulados

```python
kmeans.fit(iris.data)
```

Retirando as informações descobertas por ele (informações dos clusters + pertencimento de cada instância)

```python
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

Plotando um gráfico do resultado

```python
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=labels)
plt.scatter(centers[:, 0], centers[:, 1], marker='*', s=200, c='#050505')
plt.show()
```
